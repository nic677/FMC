```python
import numpy as np 
import matplotlib.pyplot as plt
from cycler import cycler


plt.rcParams["figure.figsize"] = (13,7)

plt.rcParams["lines.linewidth"] = 1.5
plt.rcParams["xtick.labelsize"] = "x-large"
plt.rcParams["ytick.labelsize"] = "x-large"
plt.rcParams["axes.linewidth"]  = 1.6
plt.rcParams["axes.grid"] = "True"
plt.rcParams["grid.alpha"] = 0.35
# plt.rcParams["figure.dpi"] = 250
# plt.rcParams["axes.alpha"] = 0.3
plt.rcParams["axes.prop_cycle"]= cycler('color',["#264653", "#2a9d8f", "#e9c46a","#f4a261"])
plt.rcParams["axes.titlesize"] = "x-large"
plt.rcParams["grid.linestyle"] = "--"
plt.rcParams["legend.framealpha"] = 0.45

```

```python
x = np.linspace(0,np.pi,1000)
t = 1
t_p = -t/2
E_gamma_X = -2*t*np.cos(x)-2*t-4*t_p*np.cos(x)
plt.plot(x, E_gamma_X)


E_X_M = -2*t*(-1+np.cos(x)) +4*t_p*np.cos(x)
plt.plot(x+np.pi, E_X_M)

E_M_gamma = -4*t*np.cos(np.pi-x) -4*t_p*np.cos(np.pi-x)**2
plt.plot(x+ 2*np.pi, E_M_gamma)


plt.xticks([0, np.pi, 2*np.pi, 3*np.pi], [r"$\Gamma$", "X", "M", r"$\Gamma$"]);
plt.ylabel(r"$\epsilon - \epsilon_0$", rotation=np.pi/2, labelpad=12, fontsize=15)
plt.yticks([-2,0,2,4,6],["$-2t$","$0$","$2t$","$4t$","$6t$"]);
plt.ylim(-2.1,6.1)
plt.title("Dispersion Relation")
```

```python
xx = np.linspace(-np.pi, np.pi,2000)
yy = xx*1.

xx, yy = np.meshgrid(xx, yy)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-np.pi, np.pi)

ax.plot_surface(xx, yy,-2*t*np.cos(xx)-2*t*np.cos(yy)-4*t_p*np.cos(xx)*np.cos(yy),alpha=0.65)

ax.plot(x, np.ones(len(x))*0,E_gamma_X,zdir="z",c='r')
ax.plot(np.pi*np.ones(len(x)), x, E_X_M, zdir="z", c="r")
ax.plot(np.pi-x, np.pi-x, E_M_gamma, zdir="z", c="r")
ax.view_init(20,20)
```

```python
Ek = -2*t*np.cos(xx)-2*t*np.cos(yy)-4*t_p*np.cos(xx)*np.cos(yy)


E_k_fl = Ek.flatten()
```

```python
k = np.linspace(-np.pi, np.pi,10000)
delta = .5

ea =  delta/2
eb = - delta/2
t = 1
Ep = 1/2*(ea+eb + np.sqrt((ea-eb)**2 + 16*t**2*np.cos(k/2)**2))
Em = 1/2*(ea+eb - np.sqrt((ea-eb)**2 + 16*t**2*np.cos(k/2)**2))


plt.plot(k[5000:] - 2*np.pi, Ep[5000:])
plt.plot(k[:5000] + 2*np.pi, Ep[:5000])
plt.plot(k, Em)
plt.tight_layout()
ea = 0
eb = 0
t = 1
Ep = 1/2*(ea+eb + np.sqrt((ea-eb)**2 + 16*t**2*np.cos(k/2)**2))
Em = 1/2*(ea+eb - np.sqrt((ea-eb)**2 + 16*t**2*np.cos(k/2)**2))
plt.plot(k[5000:] - 2*np.pi, Ep[5000:],'k--')
plt.plot(k[:5000] + 2*np.pi, Ep[:5000],'k--')
plt.plot(k, Em, 'k--')

plt.plot([-np.pi-0.5, -np.pi-0.5],[-delta/2, delta/2], c='k')
plt.plot([-np.pi-0.6,-np.pi-0.4],[delta/2, delta/2],c='k')
plt.plot([-np.pi-0.6,-np.pi-0.4],[-delta/2, -delta/2],c='k')
         
plt.plot([np.pi+0.5, +np.pi+0.5],[-delta/2, delta/2], c='k')
plt.plot([np.pi+0.4,np.pi+0.6],[delta/2, delta/2],c='k')
plt.plot([np.pi+0.4,np.pi+0.6],[-delta/2, -delta/2],c='k')
plt.text(-np.pi-1, -0.05, r'$\Delta$', fontsize=15 )
```
